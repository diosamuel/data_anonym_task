```python
import pandas as pd

karyawan = pd.read_csv("https://raw.githubusercontent.com/diosamuel/data_anonym_task/refs/heads/main/dataset/dataset_karyawan.csv")
pasien = pd.read_csv("https://raw.githubusercontent.com/diosamuel/data_anonym_task/refs/heads/main/dataset/dataset_pasien.csv")
siswa = pd.read_csv("https://raw.githubusercontent.com/diosamuel/data_anonym_task/refs/heads/main/dataset/dataset_siswa.csv")
```

## Supression Kolom Nama dan Generalisasi kota dari Alamat

```python
karyawan = karyawan.drop(columns=['Nama'])
karyawan['Alamat'] = karyawan['Alamat'].apply(lambda x: str(x).split(',')[-2].strip() if len(str(x).split(',')) >= 2 else None)
karyawan
```

## Tokenization Atasan

```python
import random
alphabet = list('abcdefghijklmnopqrstuvwxyz')

random.seed(42)
shuffled_alphabet = list('abcdefghijklmnopqrstuvwxyz')
random.shuffle(shuffled_alphabet)

trans_map = {}
for i in range(len(alphabet)):
    trans_map[alphabet[i]] = shuffled_alphabet[i]
    trans_map[alphabet[i].upper()] = shuffled_alphabet[i].upper()

def apply_mas(text):
    try:
        text_str = str(text)
        result = ""
        for char in text_str:
            result += trans_map.get(char, char)
        return result
    except Exception as e:
        return None

karyawan['Atasan'] = karyawan['Atasan'].apply(apply_mas)
karyawan
```

## Masking Email dan Nomor HP

```python
import re

def mask_email(email):

    try:
        return re.sub(r'^(.)[^@]*', r'\1***', str(email))
    except Exception as e:
        return None

def mask_phone(phone):

    try:
        s = re.sub(r'\D', '', str(phone))
        length = len(s)
        if length > 8:
            return s[:4] + 'X' * (length - 8) + s[-4:]
        elif length > 0:
            return s[0] + 'X' * (length - 1)
        else:
            return None

    except Exception as e:
        return None

karyawan['Email'] = karyawan['Email'].apply(mask_email)
karyawan['No_HP'] = karyawan['No_HP'].apply(mask_phone)
karyawan
```

## Numerik Variance Kolom Gaji

```python
import numpy as np

variance = np.random.uniform(-0.05, 0.05, size=len(karyawan))

karyawan['Gaji'] = (karyawan['Gaji'] * (1 + variance)).astype(int)
karyawan
```

## Generalization bulan dan tahun pada kolom Tanggal Masuk

```python
karyawan['Tanggal_Masuk'] = pd.to_datetime(karyawan['Tanggal_Masuk'], errors='coerce')

karyawan['Tanggal_Masuk'] = karyawan['Tanggal_Masuk'].dt.strftime('%m-%Y')
karyawan
```

# Masking kolom NIK

```python
def mask_nik(nik_str):
    try:
        s = str(nik_str)
        length = len(s)
        if length > 8:
            return s[:4] + 'X' * (length - 8) + s[-4:]
        elif length > 4:
            return s[:2] + 'X' * (length - 4) + s[-2:]
        else:
            return 'X' * length

    except Exception as e:
        return None

karyawan['NIK'] = karyawan['NIK'].apply(lambda x: '{:.0f}'.format(x) if isinstance(x, (int, float)) else str(x))
karyawan['NIK'] = karyawan['NIK'].apply(mask_nik)
karyawan
```
